```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('uk_accident_dataset.csv')
```

```python
df.head()
```

```python
X = df.drop('Number_of_Casualties', axis = 1)
y = df['Number_of_Casualties']
from sklearn.model_selection import StratifiedShuffleSplit
sss = StratifiedShuffleSplit(n_splits=1, test_size=0.3)
sss.split(X,y)
```

```python
for train_index, test_index in sss.split(X,y):
    X_train, y_train = X.iloc[train_index], y.iloc[train_index]
    X_test, y_test = X.iloc[test_index], y.iloc[test_index]
```

```python
X_train1 = X_train.copy()
```

```python
X_train1.head()
```

```python
for col in X_train.columns:
    print('feature:',col)
    print('data type:',X_train[col].dtype)
    print(X_train[col].value_counts()/X_train.shape[0]*100)
    print('\n')
```

```python
def delete_cols(data):
    data.drop(['2nd_Road_Number','country','Accident_ID'],axis = 1,inplace = True)
    return data
```

```python
delete_cols(X_train1)
```

```python
def cyclic_encoding(data):
    data['Time'] = pd.to_datetime(data['Time']).apply(lambda t: t.hour)
    imp_value = data['Time'].value_counts().idxmax()
    data['Time'] = data['Time'].fillna(imp_value)
    
    data['Month'] = pd.to_datetime(data['Date']).apply(lambda t: t.month)
    
    data['Day_of_Week_Sin'] = (np.sin(data['Day_of_Week']*2*np.pi/7)).astype('float16')
    data['Day_of_Week_Cos'] = (np.cos(data['Day_of_Week']*2*np.pi/7)).astype('float16')
    data['Time_Sin'] = (np.sin(data['Time']*2*np.pi/24)).astype('float16')
    data['Time_Cos'] = (np.cos(data['Time']*2*np.pi/24)).astype('float16')
    data['Month_Sin'] = (np.sin(data['Month']*2*np.pi/12)).astype('float16')
    data['Month_Cos'] = (np.cos(data['Month']*2*np.pi/12)).astype('float16')
    data.drop(['Time','Date','Day_of_Week','Month'],axis = 1, inplace = True)
    return data
```

```python
cyclic_encoding(X_train1)
```

```python
def categories_reduction(data):
    data['Pedestrian_Crossing-Human_Control'] = data['Pedestrian_Crossing-Human_Control'].replace(['Control by other authorised person','Control by school crossing patrol'],'others')
    data['Pedestrian_Crossing-Physical_Facilities'] = data['Pedestrian_Crossing-Physical_Facilities'].replace(['non-junction pedestrian crossing','Central refuge','Footbridge or subway'],'others')
    data['Light_Conditions'] = data['Light_Conditions'].apply(lambda w: w.split()[0])
    data['Weather_Conditions'] = data['Weather_Conditions'].replace(['Snowing without high winds','Fog or mist','Snowing with high winds','Unknown'],'Other')
    data['Road_Surface_Conditions'] = data['Road_Surface_Conditions'].fillna('Dry')
    data['Road_Surface_Conditions'] = data['Road_Surface_Conditions'].replace(['Frost/Ice','Snow','Flood (Over 3cm of water)'],'other')
    data['Special_Conditions_at_Site'] = data['Special_Conditions_at_Site'].fillna('None')
    data['Special_Conditions_at_Site'] = data['Special_Conditions_at_Site'].replace(['Ol or diesel','Road surface defective','Mud','Auto traffic singal out','Permanent sign or marking defective or obscured','Auto traffic signal partly defective'],'other')
    data['Carriageway_Hazards'] = data['Carriageway_Hazards'].replace(['Other object in carriageway','Any animal (except a ridden horse)','Pedestrian in carriageway (not injured)','Involvement with previous accident','Dislodged vehicle load in carriageway'],'other')
    return data
```

```python
categories_reduction(X_train1)
```

```python
import re
```

```python
def feat_transform(data):
    data['Local_Authority_(Highway)'] = data['Local_Authority_(Highway)'].apply(lambda x:x[0])
    data['postcode'] = data['postcode'].apply(lambda code: re.findall(r'^\D+',code)[0])
    return data
```

```python
feat_transform(X_train1)
```

```python
Xy_train1 = pd.concat([X_train1,y_train],axis = 1)
```

```python
dict1 = Xy_train1.groupby('postcode')['Number_of_Casualties'].mean().to_dict()
dict2 = Xy_train1.groupby('Local_Authority_(District)')['Number_of_Casualties'].mean().to_dict()
dict3 = Xy_train1.groupby('1st_Road_Number')['Number_of_Casualties'].mean().to_dict()
```

```python
def mean_encoding(data):
    data['postcode'] = data['postcode'].map(dict1)
    data['Local_Authority_(District)'] = data['Local_Authority_(District)'].map(dict2)
    data['1st_Road_Number'] = data['1st_Road_Number'].map(dict3)
    data.fillna(0)
    return data
```

```python
 mean_encoding(X_train1)
```

```python
from sklearn.preprocessing import FunctionTransformer
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV
from lightgbm import LGBMClassifier
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
cat_cols = ['Pedestrian_Crossing-Human_Control','Pedestrian_Crossing-Physical_Facilities','Light_Conditions','Weather_Conditions','Road_Surface_Conditions','Special_Conditions_at_Site','Carriageway_Hazards','Local_Authority_(Highway)','Number_of_Vehicles','1st_Road_Class','Road_Type','2nd_Road_Class','Urban_or_Rural_Area','Did_Police_Officer_Attend_Scene_of_Accident','state']
```

```python
encoding = ColumnTransformer([('cat_encoding', OneHotEncoder(sparse=False,handle_unknown='ignore'),cat_cols)],remainder='passthrough')
```

```python
f1 = FunctionTransformer(delete_cols)
f2 = FunctionTransformer(cyclic_encoding)
f3 = FunctionTransformer(categories_reduction)
f4 = FunctionTransformer(feat_transform)
f5 = FunctionTransformer(mean_encoding)
```

```python
model = LGBMClassifier()
```

```python
grid_param = {'n_estimators':[100,250,500],'learning_rate':[1,0.1,0.3]}
```

```python
GScv = GridSearchCV(model,param_grid=grid_param,cv=3,scoring='f1',n_jobs = -1)
```

```python
pipe = make_pipeline(f1,f2,f3,f4,f5,encoding,GScv)
```

```python
pipe.fit(X_train, y_train)
```

```python
pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
accuracy_score(y_test,pred)
```

```python
mean_absolute_error(y_test,pred)
```
